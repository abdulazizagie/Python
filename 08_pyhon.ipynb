# For LOOP
definitions:
    - A loop is a programming structure that allows code to be executed repeatedly.<br>
    - It is often used to repeat a block of code a certain number of times.<br>
    - For loop: Executes a block of code a certain number of times.<br>
    - Syntax:<br>
    ```python
    for variable in iterable:
    # code block
    ```.<br>
    - Example:
    ```python
    for i in range(5):
        print(i)
    ```
