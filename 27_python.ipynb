## Tkinter
definitions:
- Tkinter is a standard GUI library in Python for creating graphical user interfaces.
- It provides a platform independent way of creating graphical user interfaces.
- It is a thin wrapper around Tcl/Tk, allowing Python programs to create graphical applications.
- Tkinter is a cross-platform GUI toolkit, meaning it works on Windows, macOS, and Unix/Linux.
- Tkinter is widely used for creating GUI applications in Python.


```python
from tkinter import *
from PIL  import ImageTk,Image
from tkinter import messagebox
```

```python
root = Tk()

def handle_login():
    email = email_input.get()
    password = password_input.get()
    
    if email == "abdulaziz71655@gmail.com" and password == '123':
        messagebox.showinfo('Yayyy',"Login successful")
    
    else:
        messagebox.showerror('Error','Login Faild')


root.title("Login From")
root.iconbitmap('artificial-intelligence-icon-png-14771-Windows.ico')

# root.minsize(400,400)
# root.maxsize(400,400)
root.geometry('350x500')

root.configure(background='#0a0a0a')
img = Image.open('aaa.png')
resized_ing = img.resize((350,140))
img = ImageTk.PhotoImage(resized_ing)

img_Label = Label(root,image=img)
img_Label.pack(pady=(10,10))


text_lable = Label(root,text='Flipkart' , fg = "white",bg = '#0a0a0a')
text_lable.pack()
text_lable.config(font=("Verdana",24))


email_lable = Label(root,text="Enter Email", fg= 'white' ,bg='#0a0a0a')
email_lable.pack(pady=(20,5))
email_lable.config(font=("verdana",12))


email_input = Entry(root, width=50)
email_input.pack(ipady=6,pady=(1,15))

password_lable = Label(root,text="Enter Password" , fg='White',bg='#0a0a0a')
password_lable.pack(pady=(20,5))
password_lable.config(font=("verdana",12))

password_input = Entry(root , width=50)
password_input.pack(ipady=6,pady=(1,15))

Login_btn = Button(root , text="Login Here" , bg= "white" ,fg='black',width=30,height=3 , command=handle_login)
Login_btn.pack(pady=(10,20))
Login_btn.config(font=('vredana',10))

root.mainloop()
```
