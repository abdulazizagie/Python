
### While loop

Type of loop

1. While loop
 - It continues to loop as long as a given condition is true.
 - It is used when we don't know in advance how many times we need to iterate.
 - Syntax: while condition:
  - Body of loop
  - Indentation is used to define the body of loop.
  - Indentation is necessary in Python.
  - The loop continues until the condition becomes false.
  - Example:
  ```python
  while condition:
    print(i)
    i += 1
  print("Loop ended")
```

#### Guessing-game

````python
import random
jackpot = random.randint(1,100)
print(jackpot)
````

````python

counter = 0
guess = int(input("Enter the number"))
while jackpot != guess:
    if jackpot < guess:
        print("Higher")
    else :
        print("Lower")
    counter+=1
    guess = int(input("Enter the number"))

print("you took",counter,"attempts")
    
````
