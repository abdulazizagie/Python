# Recurison
definitions:<br>

- recursion: a function calling itself as a subroutine.<br>
- base case: a condition that terminates the recursion.<br>
- recursive function: a function that calls itself as a subroutine.<br>
- recursive algorithm: an algorithm that uses recursion to solve a problem.<br>
- tail recursion: a recursive algorithm where the recursive call is the last operation in the function.<br>
- recursive approach: a method of solving a problem by breaking it down into smaller, more manageable subproblems, solving each subproblem recursively, and combining the solutions to form the overall solution.<br>
- iterative approach: a method of solving a problem by using iteration to solve the subproblems, combining the solutions to form the overall solution.<br>
- stack overflow: an error in a recursive algorithm where the maximum call stack size is exceeded.<br>
- recursive function calls: the number of function calls made during the execution of a recursive algorithm.<br>
- space complexity: the amount of memory required by a recursive algorithm to solve a problem.<br>
- time complexity: the amount of time required by a recursive algorithm to solve a problem.<br>
- divide and conquer algorithm: an algorithm that breaks down a problem into smaller subproblems, solves each subproblem recursively, and combines the solutions to form the overall solution.<br>

```python
def multiply(a,b):
    
    result = 0
    for i in range(b):
        result = result+a
        
    print(result)
    
multiply(3,4)
```

```python
def mul(a,b):
     
    if b == 1:
        return a
    else:
        return a + mul(a,b-1)
    
print(mul(5,6))
```

# Factorial

```python
def fact(number):
    
    if number ==1:
        return 
    else:
        return number * fact(number-1)
print(5)
```

# Palindrome

```python
def palin(text):
    if len(text)<=1:
        print("palindrome")
    else:
        if (text[0]==text[-1]):
            palin(text[1:-1])
        else:
            return 'Not a Palin'
print(palin('madam'))
print(palin('malayalam'))
palin('python')
```

```python
def fib(m):
    
    if m==0 or m==1:
        return 1
    else:
        return fib(m-1)+ fib(m-2)
    
print(fib(12))
```

```python
import time
def memo(m,d):
    
    if m in d:
        return d[m]
    else:
        d[m] = memo(m-1,d) + memo(m-2,d)
        return d[m]
strat = time.time()
d = {0:1,1:1}
print(time.time()-strat)
memo(48,d)
print(d)
```

```python
def nest_list(lst):
    if len(lst) == 0:
        return [[]]
    else:
        sub_result = nest_list(lst[:-1])
        return sub_result + [[lst[-1]]]
```

```python
input_list = [1, 2]
result = nest_list(input_list)
print(result)
```
