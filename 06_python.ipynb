# Literals
Literals is a raw data given in a variable in python  there are various type of literals they are as follows.

1. Numaric Literals 
2. String Literals
3. Boolen Litterals
4. Special Literals

```python
# Numerical Literals

a = 0b1010 # Binary Literals
b =  100 #Decimal Literals
c = 0o310 #octal Literals
d = 0x12c#Hexadecimal Literal

# Float Literal
float_1 = 10.5
float_2 = 1.5e2
float_3 = 1.5e-3

# complex Literals
x =0+3.14j

print(a,b,c,d)
print(float_1,float_2,float_3)
print(x,x.imag,x.real)
```

```python
# String Literals

string = 'this is pyhton'
strings = "this is python"
char = "C"
multiling_str = """this is multiling string woth more than one line code"""
unicode =U"\U0001f600\U0001f606\U0001F923"
raw_srt = r"raw \n string"

print(string)
print(strings)
print(char)
print(multiling_str)
print(unicode)
print(raw_srt)
```

```python
# Boolen Literala 
a = True +4
b = False+10

print(a)
print(b)
```
