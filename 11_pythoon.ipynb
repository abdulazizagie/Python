## what is Break and Continue and pass in for loop.
### Break:
1. It is used to exit the loop immediately.
2. When the loop encounters a break statement, it jumps out of the loop and continues with the next iteration.
### Continue:
1. It is used to skip the current iteration and move on to the next one.
2. When the loop encounters a continue statement, it jumps to the next iteration without executing the remaining statements in the current iteration.
### Pass:
1. It is used to skip the current iteration and continue with the next one.
2. When the loop encounters a pass statement, it does nothing and continues with the next iteration.



```python
for i in range(1,11):
    if i == 5:
        break
    print(i)
```

```python
for i in range(1,11):
    if i==5:
        continue
    print(i)
```

```python
for i in range(1,11):
    pass
```
