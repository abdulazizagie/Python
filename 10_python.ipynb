```python
# rang function
print(list(range(1,11)))
print(list(range(5)))
print(list(range(1,11,2)))
print(list(range(10,0,-1)))

for i in range(1,11,2):
    print(i)
    
for i in "Pakistan":
    print(i)

# list
for i in [1,2,3,4,5]:
    print(i)
# set
for i in {1,2,3,4,5}:
    print(i)
```

    - print()

```python
rows = 5#int(input("Enter the number of row"))
for i in range(rows):
    for i in range(i+1):
        print("*",end=" ")
    print("")
```

```python
rows = 6
for i in range(rows):
    for j in range(i,rows):
        print("*",end=" ")
    print("")
        
```

```python
rows = 5

for row in range(rows):
    # for i in range(row+1):
        # print(" ",end="")
    for j in range(row+1):
        print("*" , end="")
    print(" ")
```

```python
n = 5

for i in range(n):
    
    for j in range(i,n):
        print(" ",end="")
        
    for j in range(i+1):
        print("*",end="")
    
    print("")
```

```python
n = 5

for i in range(n):
    
    for i in range(i+1):
        print(" ",end="")
    
    for j in range(i,n):
        print("*" , end="")
        
    print()
```

```python
n = 5

for i in range(n):
    
    for j in range(i,n):
        print(" ",end="")
    
    for j in range(i):
        print("*" , end= "")
    
    for j in range(i+1):
        print("*" , end="")
    
    for j in range(i,n+3):
        print(" ",end="")
    
    for j in range(i+1):
        print("*" , end= "")
    
    # for j in range(i+1):
        # print("*" , end="")
    print()
```

```python
n = 6

for i in range(n):
    
    for j in range(i+1):
        print(" ",end="")
    
    for j in range(i,n-1):
        print("*",end="")
        
    for i in range(i,n):
        print("*" , end="")
    
    print()
```

```python

# upper part
n = 5

for i in range(n-1):
    
    for j in range(i,n):
        print(" ",end=" ")
    
    for j in range(i):
        print("*",end=" ")
    
    for i in range(i+1):
        print("*", end=" ")
        
    print()

# lower part
for i in range(n):        
        for j in range(i+1):
            print(" ",end=" ")
            
        for j in range(i,n-1):
            print("*",end=" ")
            
        for j in range(i,n):
            print("*",end=" ")
            
        print()
```
