# Operators

operators are used to perfrom operstions on nariables . python has the following operators:

1. Arithmetic operators 
    These operators are used to perform arithmetic operations on numeric values.
    + : addition. <br>
    - : subtraction. <br>
    * : multiplication. <br>
    / : division.<br>
    % : modulus.<br>
2. Comparison operations
    These operators are used to compare two values and determine their relationship.<br>
    < : less than.<br>
    > : greater than.<br>
    == : equal to.<br>
    <= : less than or equal to.<br>
    >= : greater than or equal to.<br>
    != : not equal to.<br>
    is : checks if two variables refer to the same object.<br>
    is not : checks if two variables do not refer to the same object.<br>
    issubclass(class1, class2) : checks if class1 is a subclass of class2.<br>
    isinstance(obj, class) : checks if obj is an instance of class.<br>
    hasattr(obj, name) : checks if obj has an attribute named name.<br>   
3. Logical operators
    These operators are used to combine boolean values.<br>
    and : logical and.<br>
    or : logical or.<br>
    not : logical not.<br>

    Note: Some operators are not supported for all data types. For example, the modulus operator (%) is not supported for floating-point numbers.<br>
4. Bitwise operators
    These operators are used to perform bitwise operations on numbers.<br>
    & : bitwise AND.<br>
    | : bitwise OR.<br>
    ^ : bitwise XOR.<br>
    ~ : bitwise NOT.<br>
    << : bitwise left shift.<br>
    >> : bitwise right shift.<br>
    &lt;&lt; : bitwise left shift with sign extension.<br>
    &gt;&gt; : bitwise right shift with sign extension.<br>
    &lt;&lt;= : bitwise left shift with sign extension and assignment.<br>
    &gt;&gt;= : bitwise right shift with sign extension and assignment.<br>
    &lt;= : bitwise left shift with sign extension and assignment.<br>
    &gt;= : bitwise right shift with sign extension and assignment.<br>
    
    Note: Some operators are not supported for all data types. For example, the modulus operator (%) is not supported for floating-point numbers.<br>
5. Assignment operators
    These operators are used to assign values to variables.<br>
    = : assigns the value on the right side to the variable on the left side.<br>
    += : adds and assigns the value on the right side to the variable on the left side.<br>
    -= : subtracts and assigns the value on the right side to the variable on the left side.<br>
    *= : multiplies and assigns the value on the right side to the variable on the left side.<br>
    /= : divides and assigns the value on the right side to the variable on the left side.<br>
    %= : modulus and assigns the value on the right side to the variable.<br>

    Note: Some operators are not supported for all data types. For example, the modulus operator (%) is not supported for floating-point numbers..<br>
6. Identity operator
    These operators are used to compare the memory locations of two variables.
    is : checks if two variables refer to the same object.<br>
    is not : checks if two variables do not refer to the same object.<br>
    isinstance(obj, class) : checks if obj is an instance of class.<br>
    etc() : applies a given function to.<br>
    items() : returns a list of all items in an iterable.<br>
    sorted() : returns a sorted list from an iterable.<br>

    Note: Some operators are not supported for all data types. For example, the modulus operator (%) is not supported for floating-point numbers..<br>

7. Membershio operatores
    These operators are used to check if a value is present in a sequence.<br>
    in : checks if a value is present in a sequence.<br>
    not in : checks if a value is not present in a sequence.<br>
    all() : checks if all values in an iterable are true.<br>
    any() : checks if any value in an iterable is true.<br>
    etc() : applies a given function to.<br>
    items() : returns a list of all items in an iterable.<br>
    sorted() : returns a sorted list from an iterable.<br>
    zip_longest() : returns an iterator of tuples containing elements from the input iterables, filling in default values if the iterables are of unequal length.<br>
    reduce() : applies a given function to the items of an iterable and returns a single output.<br>
    max() : returns the largest item in an iterable.<br>
    min() : returns the smallest item in an iterable.<br>
    sum() : returns the sum of all items in an iterable.<br>
    
    Note: Some operators are not supported for all data types. For example, the modulus operator (%) is not supported for floating-point numbers..<br>
