```python
from tkinter import *
from PIL import ImageTk , Image
import os
```

```python
def rotate_image():
    global counter
    image_lable.config(image=image_array[counter%len(image_array)])
    counter +=1

counter = 1
root = Tk()
root.title("Wallpaper Viewer")

root.geometry('250x400')
root.configure(background="Black")


File = os.listdir("C:/Users/Hp/Downloads/Compressed/train/angry")
image_array = []
for file  in File:
    img = Image.open(os.path.join("C:/Users/Hp/Downloads/Compressed/train/angry",file))
    resized_img = img.resize((200,300))
    image_array.append(ImageTk.PhotoImage(resized_img))
    
image_lable = Label(root,image=image_array[0])
image_lable.pack(pady=(15,10))

next_btn  = Button(root,text='Next',bg='white' , fg = 'black' ,width=28 , height=2 , command=rotate_image)
next_btn.pack()


root.mainloop()
```
