# Lambda Funcations
definitions:
- A small anonymous function that can take any number of arguments but can only have one expression.
- Lambda functions are anonymous, meaning they don't have a name.
- They are defined using the keyword `lambda` followed by the arguments, a colon, and the expression to be executed.
- Lambda functions can be used wherever function objects are required.
- They are mainly used for small, one-time, anonymous functions.

## example

- Lambda input: argument1, argument2,... : expression
- E.g
- x = lambda x, y: x + y
- x(3, 5)


```python
x = lambda x: x**2
print(x(9))
```

```python
a = lambda x,y:x+y
print(a(4,5))
```

## Map
- L= [1,2,3,4,5,6,7]
- The map() function executes a specified function for each item in an iterable. The item is sent to the function as a parameter.


```python
L = [1,2,3,4,5,6,7]
print(map(lambda x: x*2 ,L))
list(map(lambda x: x*2,L))
```

```python
list(map(lambda x: x%2==0,L))
```

# Filter 
- The filter() function returns an iterator where the items are filtered through a function to test if the item is accepted or not.



```python
list(filter(lambda x:x>4,L))
```

```python
fruits = ['Apple','Orange','Mango', 'Guava']

list(filter(lambda fruits:'e' in fruits,fruits))
```
