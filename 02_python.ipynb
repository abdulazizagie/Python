## Type of Comment
1. There are two type of comment
    a. Single line comment: This comment is written on the same line as the code. It is used to explain a single line of code.
    b. Multi-line comment: This comment is written between /* and */ symbols. It can span multiple lines of code and can be used to explain multiple lines of code.
    ## Single-line comment example
    print("Hello, World!")  # This is a single-line comment explaining the print statement.
    ## Multi-line comment example
    """
    This is a multi-line comment.
    It spans multiple lines.
    This comment is used to explain multiple lines of code.
    """
    print("Hello, World!")  # This is another line of code that demonstrates the usage of a multi-line comment.
