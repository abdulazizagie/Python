# Threading
defintation_thread:
- A thread of execution in a program.
- A thread is a sequence of instructions that performs a particular task.
- A thread is created using the `Thread` class in Python's threading module.
- A thread runs concurrently with other threads in a program.
- Threads share the same memory space.

```python
from time import sleep , time

import threading

strat_time = time()
def something(id):
    
    print(f"Going to sleep...{id}")
    sleep(1)
    print(f"Woken up....{id}")

t1 = threading.Thread(target = something,args=[0])
t2 = threading.Thread(target= something,args=[1])

t1.start()
t2.start()

t1.join()
t2.join()


end_time = time()

print(f"Main Threading in {end_time-strat_time} second")

```

```python
from time import sleep , time

import threading

strat_time = time()

def something(id):
    
    print(f"Goning to sleep...{id}")
    sleep(1)
    print(f"Worken up...{id}")
    
threads = [ threading.Thread(target=something,args=[i]) for i in range(10)]

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()
    
end_time = time()

print(f"Main Thread Ended in {end_time - strat_time} seconds")
```

```python

balance = 200

lock = threading.Lock()

def deposit(amount , times , lock):
    global balance
    
    for _ in range(times):
        balance += amount
        
def withdraw(amount ,times,lock):
    global balance
    
    for _ in range(times):
        balance -= amount
        
deposit_thread = threading.Thread(target=deposit,args=[1,10000,lock])
withdraw_thread = threading.Thread(target=withdraw,args=[1,10000,lock])


deposit_thread.start()
withdraw_thread.start()

deposit_thread.join()
withdraw_thread.join()


print(balance)
```
