# What is Funcation
definitions:<br>

- A function is a block of code that performs a specific task.<br>
- Functions help to organize and modularize code, making it easier to read, understand, and maintain.<br>
- Functions can be defined using def keyword.<br>
- Functions can take input parameters, perform some operations, and return a result.<br>
- Functions can be nested, allowing for a more complex structure.<br>
- Functions can also be used as variables, allowing them to be passed as arguments to other functions.<br>
- Functions can also be defined using lambda expressions.<br>
- Functions can also be defined using annotations.<br>
- Functions can also be defined using type hints.<br>
- Functions can also be defined using custom data structures.<br>
- Functions can also be defined using classes and objects.<br>
- Functions can also be defined using generators.<br>
- Functions can also be defined using asynchronous programming.<br>
- Functions can also be defined using decorator functions.<br>

```python
def power(a,b): # is a parameter
    return a**b

print(power(2,3))

print(power(5,6))
```

```python
def power(a=1,b=2): # is a bydefult arguments
    return a**b

print(power(2))
print(power(2,3))
```

```python
def power(a=1,b=2):
    return a**b

print(power(b=2,a=3))
```
