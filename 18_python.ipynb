# Mutability

1. Mutability reders to the ability to change or edit data in it's memory location

### Immutable Data Types 
1. String
2. int
3. Float
4. Bool
5. Complex
6. Tuple


### Mutable Data Types

1. List
2. Dictionary
3. Sets

```python
L = [1,2,3]
print(L)
print(id(1))
print(id(L[0]))
print(id(2))
print(id(L[1]))
```

```python
L[2] = 1
print(L)
print(id(L[2]))
```

```python
L = [1,2,3,[4,5]]
print(L)
print(id(L[3]))
print(id(L[3][1]))
```

```python
a = "Hello"
print(id(a))

a = a + " world"
print(id(a))
```

```python
a = (1,2,3)
print(id(a))

a = a+ (5,6)
print(id(a))
```

```python
L = [1,2,3]
print(id(L))
L.append(4)
print(L)
print(id(L))
```

```python
L = [1,2,3]
L1 = L
print(id(L))
print(L)
print(id(L1))
print(L1)

L1.append(4)
print(L1)
print(L)
```
