# Tuples
definitions:
1. Tuples are immutable collection of items, i.e., cannot be changed once created.
2. They are enclosed in parentheses ( ) or square brackets [ ]. 
3. Tuples are defined using comma separated values.
4. Tuples are faster than lists in Python due to the immutability.
5. Tuples are used to group related data items and can be nested.

## Operations in tuples:
1. Indexing: Access elements using index.
2. Slicing: Extract subsets of elements.
3. Concatenation: Combine tuples.
4. Repetition: Repeat tuples.
5. Membership: Check if an element exists in a tuple.
6. Length: Get the number of elements in a tuple.

## Functions on tuples:
1. count()
2. index()
3. len()
4. max()
5. min()
6. sum()

## Nested tuples:
1. Nested tuples can contain other tuples.
2. Access elements from nested tuples using multiple index values.

## Examples:
1. Create a tuple with 5 elements.

3. Access the last three elements.
4. Access the first element of the nested tuple.
5. Create a heterogeneous tuple.
6. Create a nested tuple.

8. Get the sum of all elements in a tuple.
9. Get the maximum and minimum elements in a tuple.
10. Check if an element exists in a tuple.
11. Concatenate tuples.
12. Repeat a tuple.
