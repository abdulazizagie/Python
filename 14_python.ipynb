# Accessing substrings from a string

```python
# Concept of Indexing
c = 'Hello'
print(c)
```

```python
print(c[1:3])
```

```python
# Type of indexing
# postive indexing
# nagative indexing
print(c[-3])
```

```python
# Slicing
c = 'Hello world'
print(c)
```

```python
print(c[0:5])
```

```python
print(c[0:8:3])
```

```python
print(c[-5:-1:2])
```

```python
print(c[::-1])
```

```python
print(c[-1:-5:-1])
```
